```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df_data = pd.read_csv('https://raw.githubusercontent.com/scikit-learn/scikit-learn/main/sklearn/datasets/data/boston_house_prices.csv', header=1)
```

```python
df_data
```

```python
def calculate(pair1,pair2):
    X = df_data[[pair1]]
    y = df_data[pair2]
    model = LinearRegression()
    model.fit(X, y)
    corr_coef = df_data[pair1].corr(df_data[pair2])
    print(" ")
    print("Chart: ")
    plt.scatter(X, y)
    plt.plot(X, model.predict(X), color='red')
    plt.title(fr'{pair1} vs {pair2} (corr={corr_coef:.2f})')
    plt.xlabel(pair1)
    plt.ylabel(pair2)
    plt.show()
    print(f'Correlation: {corr_coef:.2f}')
```

```python
pairs = [("CRIM", "ZN"), ("CRIM", "INDUS"), ("CRIM", "RM"), ("AGE", "RM"), ("CRIM", "MEDV"), ("RAD", "MEDV"), ("AGE", "CRIM"), ("TAX", "AGE"), ("PTRATIO", "AGE"), ("AGE", "DIS")]
```

```python
for result in pairs:
    calculate(result[0],result[1])
```
